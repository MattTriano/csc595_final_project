# Load data and explore missingness

```python
wb_health = pd.read_csv('datasets/health_data_wb.csv')
wb_health.head(2)
```

```python
# 2016 data is mostly missing 
len(wb_health[(wb_health['year']==2016) & (wb_health['indicator']==0)])/len(wb_health[wb_health['year']==2016])
```

```python
# missingness in nutrition-related variables 
nutrition_var =['Depth of the food deficit (kilocalories per person per day)',
                'People using basic sanitation services (% of population)',
                'People using safely managed sanitation services (% of population)',
                'Prevalence of overweight, weight for height, female (% of children under 5)',
                'Prevalence of overweight, weight for height, male (% of children under 5)',
                'Prevalence of undernourishment (% of population)',
                'Prevalence of underweight, weight for age (% of children under 5)',
                'Prevalence of underweight, weight for age, female (% of children under 5)',
                'Prevalence of underweight, weight for age, male (% of children under 5)']
for i in nutrition_var:
    print(i, len(wb_health[(wb_health['indicator_name']==i) & (wb_health['indicator']!=0)])/len(wb_health[wb_health['indicator_name']==i]))
```

```python
# variables that have min_threshold data completeness
min_threshold=0.7
indicator_list=[]
data_completeness=[]
for i in wb_health.indicator_name.unique():
    if len(wb_health[(wb_health['indicator_name']==i) & (wb_health['indicator']!=0)])/len(wb_health[wb_health['indicator_name']==i])>min_threshold:
        indicator_list.append(i)
        data_completeness.append(len(wb_health[(wb_health['indicator_name']==i) & (wb_health['indicator']!=0)])/len(wb_health[wb_health['indicator_name']==i]))
#         print(i, len(wb_health[(wb_health['indicator_name']==i) & (wb_health['indicator']!=0)])/len(wb_health[wb_health['indicator_name']==i])) 
print(len(indicator_list), 'variables with at least', min_threshold, 'completeness')
pd.DataFrame(data={'indicator':indicator_list, 'data_completeness':data_completeness})[['indicator','data_completeness']]
```

```python
indicator_list
```

# select variables with sufficient data, pivot, merge with country Region and income

```python
wb_health_pivot = wb_health[wb_health['indicator_name'].isin(['Birth rate, crude (per 1,000 people)',
                                                             'Fertility rate, total (births per woman)',
                                                             'Health expenditure per capita (current US$)',
                                                             'Health expenditure, private (% of GDP)',
                                                             'Health expenditure, public (% of GDP)',
                                                             'Health expenditure, public (% of government expenditure)',
                                                             'Health expenditure, public (% of total health expenditure)',
                                                             'Health expenditure, total (% of GDP)',
                                                             'Life expectancy at birth, female (years)',
                                                             'Life expectancy at birth, male (years)',
                                                             'Life expectancy at birth, total (years)',
                                                             'Lifetime risk of maternal death (%)',
                                                             'Mortality rate, adult, female (per 1,000 female adults)',
                                                             'Mortality rate, adult, male (per 1,000 male adults)',
                                                             'Mortality rate, infant (per 1,000 live births)',
                                                             'Number of under-five deaths',
                                                             'Out-of-pocket health expenditure (% of private expenditure on health)',
                                                             'Out-of-pocket health expenditure (% of total expenditure on health)',
                                                             'People using basic sanitation services (% of population)',
                                                             'Prevalence of anemia among children (% of children under 5)',
                                                             'Prevalence of anemia among non-pregnant women (% of women ages 15-49)',
                                                             'Prevalence of anemia among pregnant women (%)',
                                                             'Prevalence of anemia among women of reproductive age (% of women ages 15-49)',
                                                             'Survival to age 65, female (% of cohort)',
                                                             'Survival to age 65, male (% of cohort)',
                                                             'Tuberculosis case detection rate (%, all forms)',
                                                             'Tuberculosis treatment success rate (% of new cases)'])].pivot_table(index=['country','year'], columns='indicator_name', values='indicator')
wb_health_pivot.reset_index(inplace=True)
wb_health_pivot.rename(columns={'Birth rate, crude (per 1,000 people)':'birth/1000',
                                'Fertility rate, total (births per woman)':'fertility/1000',
                                'Health expenditure per capita (current US$)':'health_exp/capita',
                                'Life expectancy at birth, total (years)':'life_exp',
                                 'Health expenditure, private (% of GDP)':'health_exp_private_%GDP',
                                 'Health expenditure, public (% of GDP)':'health_exp_public_%GDP',
                                 'Health expenditure, public (% of government expenditure)':'health_exp_public_%govExp',
                                 'Health expenditure, public (% of total health expenditure)':'health_exp_public_%totHealthExp',
                                 'Health expenditure, total (% of GDP)':'health_exp_%GDP',
                                 'Life expectancy at birth, female (years)':'life_exp_F',
                                 'Life expectancy at birth, male (years)':'life_exp_M',
                                 'Life expectancy at birth, total (years)':'life_exp',
                                 'Lifetime risk of maternal death (%)':'risk_maternal_death_%',
                                 'Mortality rate, adult, female (per 1,000 female adults)':'mortality/1000_adultF',
                                 'Mortality rate, adult, male (per 1,000 male adults)':'mortality/1000_adultM',
                                 'Mortality rate, infant (per 1,000 live births)':'mortality/1000_infant',
                                 'Number of under-five deaths':'death_age<5',
                                 'Out-of-pocket health expenditure (% of private expenditure on health)':'health_exp_outOfPocket_%privateExp',
                                 'Out-of-pocket health expenditure (% of total expenditure on health)':'health_exp_outOfPocket_%totExp',
                                 'People using basic sanitation services (% of population)':'sanitation_%pop',
                                 'Prevalence of anemia among children (% of children under 5)':'anemia_%age<5',
                                 'Prevalence of anemia among non-pregnant women (% of women ages 15-49)':'anemia_nonpregnant_%age15-49',
                                 'Prevalence of anemia among pregnant women (%)':'anemia_%pregnant',
                                 'Prevalence of anemia among women of reproductive age (% of women ages 15-49)':'anemia_%age15-49',
                                 'Survival to age 65, female (% of cohort)':'survival_>=age65_F',
                                 'Survival to age 65, male (% of cohort)':'survival_>=age65_M',
                                 'Tuberculosis case detection rate (%, all forms)':'tuberculosis_%detect',
                                 'Tuberculosis treatment success rate (% of new cases)':'tuberculosis_%txSuccess'
                               }, inplace=True)

wb_health_pivot.replace(0,-2,inplace=True)# replace 0s with -2 (don't want to plot 0s)
wb_health_pivot.head(2)
```

```python
countries_class=pd.read_excel('datasets/CLASS.xls', header=3, index_col='x').drop(['x.1','X'],axis=1).drop('x',axis=0)
countries_class.head(2)
```

```python
wb_health_pivot = wb_health_pivot.merge(countries_class[['Economy','Code','Region','Income group']], left_on='country', right_on='Economy', how='left')
wb_health_pivot.drop('Economy', axis=1, inplace=True)
wb_health_pivot.head(2)
```
